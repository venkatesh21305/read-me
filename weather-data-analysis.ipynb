```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Provided data
data = {
    'date': ['19790101', '19790102', '19790103', '19790104', '19790105', '19790106', '19790107', '19790108', '19790109', '19790110', '19790111'],
    'cloud_cover': [2, 6, 5, 8, 6, 5, 8, 8, 4, 7, 1],
    'sunshine': [7, 1.7, 0, 0, 2, 3.8, 0, 0.1, 5.8, 1.9, 6.8],
    'global_radiation': [52, 27, 13, 13, 29, 39, 13, 15, 50, 30, 55],
    'max_temp': [2.3, 1.6, 1.3, -0.3, 5.6, 8.3, 8.5, 5.8, 5.2, 4.9, 2.9],
    'mean_temp': [-4.1, -2.6, -2.8, -2.6, -0.8, -0.5, 1.5, 6.9, 3.7, 3.3, 2.6],
    'min_temp': [-7.5, -7.5, -7.2, -6.5, -1.4, -6.6, -5.3, 5.3, 1.6, 1.4, 0.3],
    'precipitation': [0.4, 0, 0, 0, 0, 0.7, 5.2, 0.8, 7.2, 2.1, 2.3],
    'pressure': [101900, 102530, 102050, 100840, 102250, 102780, 102520, 101870, 101170, 98700, 98960],
    'snow_depth': [9, 8, 4, 2, 1, 1, 0, 0, 0, 0, 0]
}

# Create a DataFrame from the provided data
df = pd.DataFrame(data)

# Activity 2: Data Visualization

# Line Graph: Temperature trends over time
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['max_temp'], marker='o', label='Max Temp')
plt.plot(df['date'], df['mean_temp'], marker='o', label='Mean Temp')
plt.plot(df['date'], df['min_temp'], marker='o', label='Min Temp')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature Trends Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Histogram: Distribution of precipitation values
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='precipitation', bins=10, kde=True)
plt.title('Distribution of Precipitation Values')
plt.xlabel('Precipitation')
plt.ylabel('Frequency')
plt.show()

# Heatmap: Correlation between weather variables
correlation_matrix = df[['cloud_cover', 'sunshine', 'global_radiation', 'max_temp', 'mean_temp', 'min_temp', 'precipitation', 'pressure', 'snow_depth']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
 
```
